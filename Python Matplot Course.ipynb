```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

# Figure - to create custom graphs

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])
ax.plot(x,y,c='red', alpha = 0.5, marker = '<', lw = 3.5,label = 'Sine')
ax.plot(x,z,c='green',alpha = 0.5, marker = '>',lw = 2.5,label = 'cosine')

ax.set_xlim(-5,10)
ax.set_ylim(-5,5)
ax.legend()

ax.set_xlabel('Sine')
ax.set_ylabel('Cosine')
ax.set_title('Matplotlib graph')
```

# Subplots - spliting of graphs

```python
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(10,5))

ax[0].plot(x,y)
ax[1].plot(x,z)
```
